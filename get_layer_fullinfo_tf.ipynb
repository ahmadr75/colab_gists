```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
import pandas as pd
```

```python
# Model Definition and Build
model = ResNet50(input_shape=(224, 224, 3), weights=None)
model.build(input_shape=(32, 224, 224, 3))


# Extract layers and their arguments
layer_dict = {}
for layer in model.layers:
    layer_name = layer.name
    config = layer.get_config()
    args = {k: v for k, v in config.items() if k not in ['name', 'trainable', 'dtype']}
    layer_dict[layer_name] = args

# Convert the dictionary to a DataFrame for better visualization
layer_dict_df = pd.DataFrame.from_dict(layer_dict, orient='index')

# Set pandas options to display the entire DataFrame
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)


```
